## Section 1 — Variables and Data Types

A **variable** is a named container for a value. You create one with a single `=` sign.
Python automatically determines the **type** of the value stored.

| Type | Python name | Typical chemistry use |
|------|-------------|----------------------|
| Whole number | `int` | Atomic number, number of moles (integer) |
| Decimal number | `float` | Molar mass, concentration, temperature |
| Text | `str` | Element symbol, compound name |
| True / False | `bool` | "Is the solution acidic?", convergence flags |

Good variable names are lowercase, descriptive, and use underscores for spaces:
`molar_mass_NaCl`, not `x` or `MolarMassNaCl`.


````python
# ── Example 1 ────────────────────────────────────────────────────────
atomic_number_Na  = 11           # int   — atomic number of sodium
molar_mass_NaCl   = 58.44        # float — g/mol
compound_name     = "sodium chloride"  # str
is_ionic          = True         # bool

print(f"Compound:      {compound_name}")
print(f"Atomic # (Na): {atomic_number_Na}")
print(f"Molar mass:    {molar_mass_NaCl:.2f} g/mol")
print(f"Ionic?         {is_ionic}")
print()
print(f"type(atomic_number_Na) → {type(atomic_number_Na)}")
print(f"type(molar_mass_NaCl)  → {type(molar_mass_NaCl)}")
print(f"type(compound_name)    → {type(compound_name)}")
print(f"type(is_ionic)         → {type(is_ionic)}")
````

| Operator | Meaning | Example |
|----------|---------|---------|
| `+`, `-`, `*`, `/` | add, subtract, multiply, divide | `M * n` |
| `**` | exponentiation | `10**(-7)` for 10⁻⁷ |

````python
# ── Example 2 ────────────────────────────────────────────────────────
# How many grams of NaCl to make 250 mL of 0.500 M solution?
molarity_NaCl   = 0.500    # mol/L
volume_L        = 0.250    # L
molar_mass_NaCl = 58.44    # g/mol

moles_NaCl = molarity_NaCl * volume_L
mass_NaCl  = moles_NaCl * molar_mass_NaCl

print(f"Moles of NaCl needed: {moles_NaCl:.4f} mol")
print(f"Mass of NaCl needed:  {mass_NaCl:.2f} g")
print()

# Beer–Lambert law: A = ε × l × c
epsilon    = 5500.0   # L mol⁻¹ cm⁻¹
path_len   = 1.0      # cm
conc       = 2.5e-4   # mol/L
absorbance = epsilon * path_len * conc
print(f"Absorbance A = {absorbance:.4f}")
````


````python
# ── Example 5 ────────────────────────────────────────────────────────
absorbances = [0.082, 0.155, 0.241, 0.318, 0.397, 0.461]
epsilon, path_length = 5500.0, 1.0

concentrations = []
for i, A in enumerate(absorbances):
    time_min = i * 5
    c_uM = (A / (epsilon * path_length)) * 1e6
    concentrations.append(c_uM)
    print(f"t = {time_min:2d} min   A = {A:.3f}   c = {c_uM:.2f} μM")
````

````python
# ── Example 7 ────────────────────────────────────────────────────────
import math

def ideal_gas_volume(n_moles, temp_C=25.0, pressure_atm=1.0):
    """Calculate volume (L) of an ideal gas: PV = nRT."""
    R = 0.082057   # L·atm / (mol·K)
    T = temp_C + 273.15
    return (n_moles * R * T) / pressure_atm

def pH_from_conc(H_concentration):
    """Calculate pH from [H⁺] in mol/L."""
    return -math.log10(H_concentration)

print(f"1 mol at 25°C, 1 atm: {ideal_gas_volume(1.0):.3f} L")
print(f"1 mol at  0°C, 1 atm: {ideal_gas_volume(1.0, temp_C=0.0):.3f} L  ← STP")
for H in [1e-2, 1e-7, 1e-12]:
    print(f"[H⁺] = {H:.0e}  →  pH = {pH_from_conc(H):.1f}")
````

````python
# ── Example 8b — read and find λ_max ────────────────────────────────
wavelengths, absorbances = [], []
with open("uvvis_spectrum.txt", "r") as fh:
    for line in fh:
        line = line.strip()
        if line.startswith("#") or not line:
            continue
        parts = line.split(", ")
        wavelengths.append(float(parts[0]))
        absorbances.append(float(parts[1]))

max_A      = max(absorbances)
lambda_max = wavelengths[absorbances.index(max_A)]
print(f"Read {len(wavelengths)} points.  λ_max = {lambda_max} nm  (A = {max_A:.3f})")
````

## Capstone Exercise — Enzyme Kinetics

The **Michaelis–Menten equation**: $v_0 = V_{\max} \cdot [S] \;/\; (K_m + [S])$


````python
# ── Capstone setup — run this first ──────────────────────────────────
V_max = 120.0   # μmol/min
K_m   = 25.0    # μM
substrate_concs = [1, 2, 5, 10, 15, 20, 25, 30, 40, 50, 75, 100, 150, 200, 300]

def michaelis_menten(S, V_max, K_m):
    """Return reaction rate v0 for substrate concentration S."""
    return (V_max * S) / (K_m + S)

with open("kinetics_data.txt", "w") as fh:
    fh.write("# [S] (uM), v0 (umol/min)\n")
    for S in substrate_concs:
        fh.write(f"{S}, {michaelis_menten(S, V_max, K_m):.4f}\n")
print("Written: kinetics_data.txt  (V_max=120, Km=25)")
````
