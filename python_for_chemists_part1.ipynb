## Section 1 — Variables and Data Types

A **variable** is a named container for a value. You create one with a single `=` sign.
Python automatically determines the **type** of the value stored.

The four types you will use constantly in chemistry:

| Type | Python name | Typical chemistry use |
|------|-------------|----------------------|
| Whole number | `int` | Atomic number, number of moles (integer) |
| Decimal number | `float` | Molar mass, concentration, temperature |
| Text | `str` | Element symbol, compound name |
| True / False | `bool` | "Is the solution acidic?", convergence flags |

Good variable names are lowercase, descriptive, and use underscores for spaces:
`molar_mass_NaCl`, not `x` or `MolarMassNaCl`.


````python
# ── Example 1 ────────────────────────────────────────────────────────
# Variables describing sodium chloride
atomic_number_Na  = 11           # int   — atomic number of sodium
molar_mass_NaCl   = 58.44        # float — g/mol
compound_name     = "sodium chloride"  # str
is_ionic          = True         # bool

# f-strings let you embed variables directly in printed text
# The :.2f format specifier means "show 2 decimal places"
print(f"Compound:      {compound_name}")
print(f"Atomic # (Na): {atomic_number_Na}")
print(f"Molar mass:    {molar_mass_NaCl:.2f} g/mol")
print(f"Ionic?         {is_ionic}")
print()

# type() tells you the data type Python assigned
print(f"type(atomic_number_Na) → {type(atomic_number_Na)}")
print(f"type(molar_mass_NaCl)  → {type(molar_mass_NaCl)}")
print(f"type(compound_name)    → {type(compound_name)}")
print(f"type(is_ionic)         → {type(is_ionic)}")
````

## Section 2 — Arithmetic and Unit Conversion

Python handles arithmetic with the operators you expect, plus two that are easy to miss:

| Operator | Meaning | Example |
|----------|---------|---------|
| `+`, `-`, `*`, `/` | add, subtract, multiply, divide | `M * n` |
| `**` | exponentiation (power) | `10**(-7)` for 10⁻⁷ |
| `//` | integer (floor) division | `7 // 2 → 3` |
| `%` | modulus (remainder) | `7 % 2 → 1` |

Python follows standard order of operations (PEMDAS). Use parentheses liberally —
they cost nothing and prevent mistakes.


````python
# ── Example 2 ────────────────────────────────────────────────────────
# ─── How many grams of NaCl do we need to make 250 mL of 0.500 M solution? ───
# n = C × V   (moles = molarity × volume in litres)
# m = n × M   (mass = moles × molar mass)

molarity_NaCl  = 0.500    # mol/L
volume_L       = 0.250    # L  (250 mL converted)
molar_mass_NaCl = 58.44   # g/mol

moles_NaCl = molarity_NaCl * volume_L
mass_NaCl  = moles_NaCl   * molar_mass_NaCl

print(f"Moles of NaCl needed: {moles_NaCl:.4f} mol")
print(f"Mass of NaCl needed:  {mass_NaCl:.2f} g")
print()

# ─── Beer–Lambert law: A = ε × l × c ─────────────────────────────────
epsilon  = 5500.0   # L mol⁻¹ cm⁻¹ (molar absorptivity)
path_len = 1.0      # cm
conc     = 2.5e-4   # mol/L  (scientific notation: 2.5 × 10⁻⁴)

absorbance = epsilon * path_len * conc
print(f"Absorbance A = {absorbance:.4f}  (dimensionless)")
````

## Section 3 — Strings

A **string** (`str`) is any sequence of characters enclosed in quotes.
Text from instrument reports, data files, and labels in plots will all arrive as strings.

The most useful string operations for chemists:

| Operation | What it does | Example |
|-----------|-------------|---------|
| `s.upper()` / `s.lower()` | change case | `"na".upper()` → `"NA"` |
| `s.strip()` | remove leading/trailing whitespace | `"  H2O  ".strip()` → `"H2O"` |
| `s.split()` | split on whitespace → list | `"1.23 nm".split()` → `["1.23", "nm"]` |
| `s.split(sep)` | split on a separator | `"H,C,N".split(",")` → `["H","C","N"]` |
| `s.replace(old, new)` | substitute text | `"sulphur".replace("ph","f")` |
| `sub in s` | membership test | `"H" in "H2O"` → `True` |
| `len(s)` | number of characters | `len("NaCl")` → `4` |


````python
# ── Example 3 ────────────────────────────────────────────────────────
formula = "  C6H12O6  "   # glucose, with accidental whitespace

# Clean up
clean = formula.strip()
print(f"Cleaned formula: '{clean}'")

# Does the formula contain oxygen?
print(f"Contains oxygen? {'O' in clean}")

# Split a compound data entry from a CSV-style line
data_line = "glucose, 180.16, 6, solid"
parts = data_line.split(", ")
name, molar_mass, n_carbons, phase = parts
print()
print(f"Name:       {name}")
print(f"Molar mass: {molar_mass} g/mol")   # still a string!
print(f"Carbons:    {n_carbons}")
print(f"Phase:      {phase}")
print()

# Convert the molar mass to a float for arithmetic
molar_mass_float = float(molar_mass)
print(f"Molar mass as float: {molar_mass_float:.2f}")

# Build a descriptive label using an f-string
label = f"{name.capitalize()} (M = {molar_mass_float:.1f} g/mol)"
print(f"Plot label: '{label}'")
````

| Operation | Example | Result |
|-----------|---------|--------|
| Index (0-based) | `data[0]` | first item |
| Negative index | `data[-1]` | last item |

## Section 5 — Loops

A **`for` loop** repeats a block of code once for each item in a sequence.
This is how you process an entire column of experimental data without copy-pasting.

```python
for item in collection:
    # do something with item
```

Two useful companions:
- `enumerate(collection)` — gives you both the index and the value at each step
- `zip(list1, list2)` — pairs items from two lists side by side


````python
# ── Example 5 ────────────────────────────────────────────────────────
# Convert a series of absorbances to concentrations using Beer–Lambert
# c = A / (ε × l)

absorbances = [0.082, 0.155, 0.241, 0.318, 0.397, 0.461]
epsilon      = 5500.0   # L mol⁻¹ cm⁻¹
path_length  = 1.0      # cm

print(f"{'Time (min)':<12} {'Absorbance':>12} {'Concentration (μM)':>20}")
print("-" * 46)

concentrations = []    # we'll collect results here
for i, A in enumerate(absorbances):
    time_min = i * 5        # one reading every 5 minutes
    c_mol_L  = A / (epsilon * path_length)
    c_uM     = c_mol_L * 1e6   # convert mol/L → μmol/L (μM)
    concentrations.append(c_uM)
    print(f"{time_min:<12} {A:>12.3f} {c_uM:>20.2f}")

print()
print(f"Final concentration: {concentrations[-1]:.2f} μM")
````

## Section 6 — Dictionaries

A **dictionary** (`dict`) maps *keys* to *values*. Think of it like an index card file:
you look up a key and retrieve the associated data instantly.

```python
d = {"key": value, "key2": value2}
d["key"]          # retrieve a value
d["key3"] = v3    # add a new entry
"key" in d        # True/False — is this key present?
d.keys()          # all keys
d.values()        # all values
d.items()         # all key-value pairs (for looping)
```

Dictionaries are ideal for periodic-table look-ups, spectral libraries, and any time
you want to retrieve data by name rather than by position.


````python
# ── Example 6 ────────────────────────────────────────────────────────
# A small periodic-table dictionary: symbol → properties
elements = {
    "H":  {"name": "Hydrogen",   "atomic_number": 1,  "atomic_mass": 1.008,   "electronegativity": 2.20},
    "C":  {"name": "Carbon",     "atomic_number": 6,  "atomic_mass": 12.011,  "electronegativity": 2.55},
    "N":  {"name": "Nitrogen",   "atomic_number": 7,  "atomic_mass": 14.007,  "electronegativity": 3.04},
    "O":  {"name": "Oxygen",     "atomic_number": 8,  "atomic_mass": 15.999,  "electronegativity": 3.44},
    "F":  {"name": "Fluorine",   "atomic_number": 9,  "atomic_mass": 18.998,  "electronegativity": 3.98},
    "Cl": {"name": "Chlorine",   "atomic_number": 17, "atomic_mass": 35.45,   "electronegativity": 3.16},
}

# Look up a single element
sym = "O"
props = elements[sym]
print(f"{props['name']} ({sym})")
print(f"  Atomic number:    {props['atomic_number']}")
print(f"  Atomic mass:      {props['atomic_mass']} u")
print(f"  Electronegativity:{props['electronegativity']}")
print()

# Loop over the whole table
print(f"{'Symbol':<6} {'Name':<12} {'Mass (u)':>10} {'EN':>6}")
print("-" * 38)
for sym, p in elements.items():
    print(f"{sym:<6} {p['name']:<12} {p['atomic_mass']:>10.3f} {p['electronegativity']:>6.2f}")

print()
# Which elements have EN > 3.0?
electronegative = [sym for sym, p in elements.items() if p["electronegativity"] > 3.0]
print(f"Elements with EN > 3.0: {electronegative}")
````

## Section 7 — Functions

A **function** packages a calculation so you can reuse it without retyping.
This is crucial in chemistry where the same equation is applied to dozens of samples.

```python
def function_name(parameter1, parameter2):
    """One-line description (docstring)."""
    result = ...          # do the calculation
    return result         # send the answer back
```

Good functions:
- Do **one thing** and do it well
- Have a **docstring** explaining inputs and outputs
- Use **descriptive parameter names**

Python also has **optional parameters** with default values — very handy for things like
standard temperature or path length that rarely change.


````python
# ── Example 7 ────────────────────────────────────────────────────────
import math

def ideal_gas_volume(n_moles, temp_C=25.0, pressure_atm=1.0):
    """
    Calculate the volume of an ideal gas using PV = nRT.

    Parameters
    ----------
    n_moles     : float — amount of gas (mol)
    temp_C      : float — temperature in Celsius (default 25 °C)
    pressure_atm: float — pressure in atm (default 1 atm)

    Returns
    -------
    float — volume in litres
    """
    R = 0.082057   # L·atm / (mol·K)
    T = temp_C + 273.15
    return (n_moles * R * T) / pressure_atm


def pH_from_conc(H_concentration):
    """Calculate pH from [H⁺] in mol/L."""
    return -math.log10(H_concentration)


# --- Using the functions ---
V1 = ideal_gas_volume(1.0)                    # 1 mol at 25 °C, 1 atm
V2 = ideal_gas_volume(1.0, temp_C=0.0)        # 1 mol at 0 °C (STP)
V3 = ideal_gas_volume(2.5, temp_C=100.0, pressure_atm=2.0)

print(f"1 mol at  25 °C, 1 atm: {V1:.3f} L")
print(f"1 mol at   0 °C, 1 atm: {V2:.3f} L  ← molar volume at STP")
print(f"2.5 mol at 100 °C, 2 atm: {V3:.3f} L")
print()

for H_conc in [1e-2, 1e-7, 1e-12]:
    print(f"[H⁺] = {H_conc:.0e} mol/L  →  pH = {pH_from_conc(H_conc):.1f}")
````

## Section 8 — Reading and Writing Files

Almost all instrument data arrives as a text file. Python's built-in `open()` function
lets you read any plain-text file line by line.

```python
# Reading
with open("data.txt", "r") as fh:
    lines = fh.readlines()   # list of strings, one per line

# Writing
with open("results.txt", "w") as fh:
    fh.write("some text\n")
```

The `with` statement ensures the file is closed automatically — always use it.

Each line read by `readlines()` ends with a newline character `\n`.
Call `.strip()` to remove it before processing.

We will first **write** a small simulated data file, then **read** it back and analyse it.


````python
# ── Example 8a — write a simulated UV-Vis spectrum ────────────────────
# Format: wavelength (nm), absorbance  — one pair per line
spectrum_data = [
    (300, 0.021), (310, 0.045), (320, 0.098), (330, 0.187),
    (340, 0.312), (350, 0.458), (360, 0.581), (370, 0.634),
    (380, 0.589), (390, 0.492), (400, 0.371), (410, 0.261),
    (420, 0.163), (430, 0.095), (440, 0.052), (450, 0.028),
]

with open("uvvis_spectrum.txt", "w") as fh:
    fh.write("# Wavelength(nm), Absorbance\n")   # comment/header line
    for wl, A in spectrum_data:
        fh.write(f"{wl}, {A:.3f}\n")

print("File written: uvvis_spectrum.txt")
````

````python
# ── Example 8b — read and analyse the spectrum ───────────────────────
wavelengths  = []
absorbances  = []

with open("uvvis_spectrum.txt", "r") as fh:
    for line in fh:
        line = line.strip()
        if line.startswith("#") or line == "":  # skip comments & blank lines
            continue
        parts = line.split(", ")
        wavelengths.append(float(parts[0]))
        absorbances.append(float(parts[1]))

print(f"Data points read: {len(wavelengths)}")
print()

# Find the absorption maximum (λ_max)
max_A  = max(absorbances)
idx    = absorbances.index(max_A)
lambda_max = wavelengths[idx]

print(f"λ_max = {lambda_max} nm  (A = {max_A:.3f})")
print()

# Print the full spectrum in a table
print(f"{'λ (nm)':>8} {'A':>8}")
print("-" * 18)
for wl, A in zip(wavelengths, absorbances):
    marker = " ← max" if wl == lambda_max else ""
    print(f"{wl:>8.0f} {A:>8.3f}{marker}")
````

## Capstone Exercise — Putting It All Together

You have just run an **enzyme kinetics** experiment. The data file below contains
substrate concentration [S] (in μM) and the corresponding initial reaction rate v₀
(in μmol/min).

The **Michaelis–Menten equation** describes enzyme kinetics:

$$v_0 = \frac{V_{\max} \cdot [S]}{K_m + [S]}$$

where V_max is the maximum rate and K_m is the Michaelis constant.

In this exercise you will:
1. Write and read the data file
2. Compute v₀ from the Michaelis–Menten equation using known parameters
3. Find the [S] that gives a rate closest to V_max / 2 (which should be ≈ K_m)
4. Identify the substrate concentration range where v₀ is within 10% of V_max (near-saturation)


````python
# ── Capstone — Part A: create the data file ───────────────────────────
# Known parameters
V_max = 120.0   # μmol/min
K_m   = 25.0    # μM

# Substrate concentrations to test (μM)
substrate_concs = [1, 2, 5, 10, 15, 20, 25, 30, 40, 50, 75, 100, 150, 200, 300]

def michaelis_menten(S, V_max, K_m):
    """Return reaction rate v0 for substrate concentration S."""
    return (V_max * S) / (K_m + S)

# Write data file
with open("kinetics_data.txt", "w") as fh:
    fh.write("# [S] (uM), v0 (umol/min)\n")
    for S in substrate_concs:
        v0 = michaelis_menten(S, V_max, K_m)
        fh.write(f"{S}, {v0:.4f}\n")

print("Written: kinetics_data.txt")
````

````python
# ── Capstone — Part B: read and analyse ──────────────────────────────
# Read the file
S_values = []
v0_values = []

with open("kinetics_data.txt", "r") as fh:
    for line in fh:
        line = line.strip()
        if line.startswith("#") or not line:
            continue
        s_str, v_str = line.split(", ")
        S_values.append(float(s_str))
        v0_values.append(float(v_str))

# Print the kinetics table
print(f"{'[S] (μM)':>10} {'v₀ (μmol/min)':>15} {'% of Vmax':>12}")
print("-" * 40)
for S, v0 in zip(S_values, v0_values):
    pct = (v0 / V_max) * 100
    print(f"{S:>10.0f} {v0:>15.4f} {pct:>12.1f}%")

print()

# Find [S] closest to half-maximal rate (should be ≈ K_m)
half_Vmax = V_max / 2
diffs = [abs(v - half_Vmax) for v in v0_values]
km_idx = diffs.index(min(diffs))
print(f"Half-maximal rate ({half_Vmax:.1f} μmol/min) is closest at [S] = {S_values[km_idx]:.0f} μM")
print(f"True K_m = {K_m} μM  ← confirms our analysis")
print()

# Find near-saturation range (v0 > 90% of Vmax)
saturated = [(S, v0) for S, v0 in zip(S_values, v0_values) if v0 >= 0.9 * V_max]
print(f"Near-saturation (v₀ ≥ 90% Vmax) starts at [S] = {saturated[0][0]:.0f} μM")
````
